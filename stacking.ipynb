```python
subm_dir = os.path.join(os.getcwd(),'submission_others')
subm_files = os.listdir(subm_dir)
subm_files
```

```python
cols = [pd.read_csv(os.path.join(subm_dir,s),index_col=0) for s in subm_files]
subm_concat = pd.concat(cols,axis=1)
col_names = list(map(lambda x: 'is_iceberg_' + str(x), range(len(subm_concat.columns))))
subm_concat.columns = subm_files
subm_concat.reset_index(inplace=True)
subm_concat.head()
```

```python
ncols = len(subm_concat.columns)
print(ncols)
```

```python
subm_concat.corr()
```

```python
lo_cutoff = 0.2
hi_cutoff = 0.8
```

```python
subm_concat['is_iceberg_max'] = subm_concat.iloc[:,1:ncols].max(axis=1)
subm_concat['is_iceberg_min'] = subm_concat.iloc[:,1:ncols].min(axis=1)
subm_concat['is_iceberg_mean'] = subm_concat.iloc[:,1:ncols].mean(axis=1)
subm_concat['is_iceberg_median'] = subm_concat.iloc[:,1:ncols].median(axis=1)
subm_concat.head()
```

```python
#min-max plus median
subm_concat['is_iceberg'] = np.where(np.all(subm_concat.iloc[:,1:ncols] > lo_cutoff,axis=1),
                                     subm_concat['is_iceberg_max'],np.where(np.all(subm_concat.iloc[:,1:ncols] < hi_cutoff,axis=1),
                                                                           subm_concat['is_iceberg_min'],
                                                                           subm_concat['is_iceberg_median']))
subm_concat.tail()
```

```python
submission= subm_concat[['index', 'is_iceberg']]
submission.columns = ['id','is_iceberg']

submission_filename = 'stack_minmax_median2.csv'
submission.to_csv(submission_filename, index=False)

FileLink(submission_filename)
```
